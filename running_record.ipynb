```python
import network3
from network3 import Network
from network3 import ReLU  ##这段代码还是没有使用ReLU激活函数
from network3 import ConvPoolLayer, FullyConnectedLayer, SoftmaxLayer
training_data, validation_data, test_data = network3.load_data_shared()
mini_batch_size = 10

net = Network([
        ConvPoolLayer(image_shape=(mini_batch_size, 1, 28, 28), 
                      filter_shape=(20, 1, 5, 5), 
                      poolsize=(2, 2)),
        FullyConnectedLayer(n_in=20*12*12, n_out=100),
        SoftmaxLayer(n_in=100, n_out=10)], mini_batch_size)
net.SGD(training_data, 60, mini_batch_size, 0.1, 
            validation_data, test_data)
##在上一次训练网络的基础上添加了卷积池化层，这层神经元将卷积和池化操作一同进行以简化代码。
##由于这两类操作通常共同存在，因此将之合并是可行的。该层的作用主要是提取图像边缘并剔除冗余信息简化输出。
##Best validation accuracy of 98.76% obtained at iteration 299999
##Corresponding test accuracy of 98.72%
```

```python
import network3
from network3 import Network
from network3 import ReLU
from network3 import ConvPoolLayer, FullyConnectedLayer, SoftmaxLayer
training_data, validation_data, test_data = network3.load_data_shared()
mini_batch_size = 10

net = Network([
        ConvPoolLayer(image_shape=(mini_batch_size, 1, 28, 28), 
                      filter_shape=(20, 1, 5, 5), 
                      poolsize=(2, 2)),
        ConvPoolLayer(image_shape=(mini_batch_size, 20, 12, 12), 
                      filter_shape=(40, 20, 5, 5), 
                      poolsize=(2, 2)),
        FullyConnectedLayer(n_in=40*4*4, n_out=100),
        SoftmaxLayer(n_in=100, n_out=10)], mini_batch_size)
net.SGD(training_data, 60, mini_batch_size, 0.1, 
            validation_data, test_data)
##这段代码额外添加了一层卷积池化神经元，相应地，其全连接层的输入张量规模也对应减小。
##Best validation accuracy of 99.06% obtained at iteration 299999
##Corresponding test accuracy of 98.89%  可以看到，两次提取边缘对训练结果有一点点微小地提升
```

```python
import network3
from network3 import Network
from network3 import ReLU
from network3 import ConvPoolLayer, FullyConnectedLayer, SoftmaxLayer
training_data, validation_data, test_data = network3.load_data_shared()
mini_batch_size = 10

net = Network([
        ConvPoolLayer(image_shape=(mini_batch_size, 1, 28, 28), 
                      filter_shape=(20, 1, 5, 5), 
                      poolsize=(2, 2), 
                      activation_fn=ReLU),  ##这段代码中，卷积池化和全连接层的激活函数都改成了ReLU函数
        ConvPoolLayer(image_shape=(mini_batch_size, 20, 12, 12), 
                      filter_shape=(40, 20, 5, 5), 
                      poolsize=(2, 2), 
                      activation_fn=ReLU),
        FullyConnectedLayer(n_in=40*4*4, n_out=100, activation_fn=ReLU),
        SoftmaxLayer(n_in=100, n_out=10)], mini_batch_size)
net.SGD(training_data, 60, mini_batch_size, 0.03, 
            validation_data, test_data, lmbda=0.1)

##ReLU给梯度下降和反向传播的效率带来了提升，避免了梯度爆炸和梯度消失问题。
#Best validation accuracy of 99.02% obtained at iteration 299999
#Corresponding test accuracy of 99.16%
#可以看出，相对于sigmoid激活函数，ReLU激活函数给网络识别性能带来的提升还是比较明显的。
```

```python
import network3
from network3 import Network
from network3 import ReLU
from network3 import ConvPoolLayer, FullyConnectedLayer, SoftmaxLayer
training_data, validation_data, test_data = network3.load_data_shared()
mini_batch_size = 10

expanded_training_data, _, _ = network3.load_data_shared(
        "../data/mnist_expanded.pkl.gz")  ##事先使用算法拓展了训练数据集的大小，载入数据重定向为拓展后的数据集

net = Network([
        ConvPoolLayer(image_shape=(mini_batch_size, 1, 28, 28), 
                      filter_shape=(20, 1, 5, 5), 
                      poolsize=(2, 2), 
                      activation_fn=ReLU),
        ConvPoolLayer(image_shape=(mini_batch_size, 20, 12, 12), 
                      filter_shape=(40, 20, 5, 5), 
                      poolsize=(2, 2), 
                      activation_fn=ReLU),
        FullyConnectedLayer(n_in=40*4*4, n_out=100, activation_fn=ReLU),
        SoftmaxLayer(n_in=100, n_out=10)], mini_batch_size)

net.SGD(expanded_training_data, 30, mini_batch_size, 0.03, 
            validation_data, test_data, lmbda=0.1)

##用于拓展训练数据集的算法是对每张图片的同一个像素点，在它的上下左右四个点，用新的像素代替原图像的像素。
##拓展后的训练集共有25万张图片，训练网络结构不变，训练结果相较于上一节代码有微小提升。
##本次实验出于加快训练速度的考虑，将训练次数降低为30个epoch（原60）
#Best validation accuracy of 99.41% obtained at iteration 649999
#Corresponding test accuracy of 99.30%
```

```python
import network3
from network3 import Network
from network3 import ReLU
from network3 import ConvPoolLayer, FullyConnectedLayer, SoftmaxLayer
training_data, validation_data, test_data = network3.load_data_shared()
mini_batch_size = 10

expanded_training_data, _, _ = network3.load_data_shared(
        "../data/mnist_expanded.pkl.gz")
net = Network([
        ConvPoolLayer(image_shape=(mini_batch_size, 1, 28, 28), 
                      filter_shape=(20, 1, 5, 5), 
                      poolsize=(2, 2), 
                      activation_fn=ReLU),
        ConvPoolLayer(image_shape=(mini_batch_size, 20, 12, 12), 
                      filter_shape=(40, 20, 5, 5), 
                      poolsize=(2, 2), 
                      activation_fn=ReLU),
        FullyConnectedLayer(n_in=40*4*4, n_out=100, activation_fn=ReLU),
        FullyConnectedLayer(n_in=100, n_out=100, activation_fn=ReLU),
        SoftmaxLayer(n_in=100, n_out=10)], mini_batch_size)
net.SGD(expanded_training_data, 30, mini_batch_size, 0.03, 
            validation_data, test_data, lmbda=0.1)

##两个全连接层带来一定的性能提升，效果不太显著。
##Best validation accuracy of 99.38% obtained at iteration 649999
##Corresponding test accuracy of 99.35%



##另外，在最后一段代码中，作者使用强正则化（dropout technique）来减轻过拟合趋势，时间所限，运行结果没有贴出。
```
